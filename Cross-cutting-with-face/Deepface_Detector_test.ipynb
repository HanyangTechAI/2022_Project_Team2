```python
from deepface import DeepFace
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import cv2
import glob
import time
from yoloface.face_detector import YoloDetector
```

## DeepFace Detector Test

```python
backends = [
  'opencv', 
  'ssd', 
  #'dlib', 
  'mtcnn', 
  'retinaface', 
  #'mediapipe'
]

face = DeepFace.detectFace(img_path = "test_image.jpg",
                           target_size = (224, 224),
                          detector_backend = backends[0], align=False
                          )
plt.imshow(face)
```

```python
def get_all_images():
    images = []
    frames = glob.glob('./dataset/frame/*')
    for frame in frames :
        for image in glob.glob(frame + '/*.jpg'):
            images.append(image)
    return images
                
def test_detector(detector):
    images = get_all_images()
    
    if detector == 'yoloface':
        yolo_detector = YoloDetector(target_size=720, gpu=-1, min_face=90)
        for image in images:
            image = np.array(Image.open(image))
            boxes, points = yolo_detector.predict(image)
    else : 
        for image in images:
            face = DeepFace.detectFace(img_path = image,
                           target_size = (224, 224),
                          detector_backend = detector, align=False
                          )
```

```python
detectors = [
  'opencv', 
  'ssd', 
  #'dlib', 
  'mtcnn', 
  'retinaface', 
   #'mediapipe',
    'yoloface'
]
print(f'총 {len(get_all_images())}개 이미지분석 시간 측정 시작')
for detector in detectors:
    print(detector + ' time')
    %time test_detector(detector)
```
